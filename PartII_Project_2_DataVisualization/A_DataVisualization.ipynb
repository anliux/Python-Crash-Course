## 15.2 绘制简单的折线图

### 绘制步骤及讲解：
    导入模块pyplot，并指定别名plt。pyplot包含很多用于生成图表的函数
    函数plot()：根据参数绘制有意义的图形
    plt.show()：打开matplotlib查看器，并显示绘制的图形。
    查看器能够缩放和导航图形，单击磁盘图标可将图形保存起来。

```python
# mpl_squares.py
import matplotlib.pyplot as plt

squares = [1, 4, 9, 16, 25]  # 创建一个列表，在其中存储一写平方数
plt.plot(squares)  # 将上面的列表传递给函数plot()，以绘制出有意义的图形
plt.show()  # 打开matplotlib查看器，并显示绘制的图形
```

### 15.2.1 修改标签文字和线条粗细
    设置线条粗细、标题、坐标轴标题及字号、以及刻度大小和刻度字号

```python
# mpl_squares.py
import matplotlib.pyplot as plt

squares = [1, 4, 9, 16, 25]
plt.plot(squares, linewidth=5)  # linewidth决定了plot()绘制的线条的粗细

# 设置图表标题，并给坐标轴加上标签
plt.title('Square Numbers', fontsize=24)  # title():给图表指定标题，fontsize参数：指定文字的大小 
plt.xlabel('Value', fontsize=14)  # xlabel，ylabel设置每条轴的标题
plt.ylabel('Square of Value', fontsize=14)

# 设置刻度标记的大小
plt.tick_params(axis='both', labelsize=14)  # both-大概是影响x和y轴的刻度，labelsize也是字号14

plt.show()
```

### 15.2.2 校正图形
    出现错误：折线图的终点4的平方为25
    原因：向plot()提供一系列数字时，它假设第一个数据点对应的x为0，而这里需要对应1
    解决：为改变这种默认行为，可以给plot()同时提供输入值和输出值
    如下代码：加入input_values，并在plt.plot()加入这个参数

```python
# mpl_squares.py
import matplotlib.pyplot as plt

# 这里做相应的改变，加入输入值和输出值，明确输入值是从1开始的
input_values = [1, 2, 3, 4, 5]
squares = [1, 4, 9, 16, 25]
plt.plot(input_values, squares, linewidth=5)

# 设置图表标题，并给坐标轴加上标签
plt.title('Square Numbers', fontsize=24) 
plt.xlabel('Value', fontsize=14)
plt.ylabel('Square of Value', fontsize=14)

# 设置刻度标记的大小
plt.tick_params(axis='both', labelsize=14)

plt.show()
```

### 15.2.3 使用scatter()绘制散点图并设置其样式
    例如：想以不同颜色表示较大或较小的数值时。
    设置各个数据点的样式，或对每个点设置同样的样式，再使用不同的样式选项重新绘制这些点以突出。
    函数scatter()：绘制单个点，并向它传递一对x和y坐标

```python
# scatter_squares.py
import matplotlib.pyplot as plt

plt.scatter(2, 4)
plt.show()
```

#### 在上面代码的基础上，设置输出样式：添加标题，给轴加标签，并确保所有文本能看清

```python
# scatter_squares.py
import matplotlib.pyplot as plt

plt.scatter(2, 4, s=200)  # s设置点的尺寸

# 设置图表标题并给坐标轴加上标签
plt.title('Square Number', fontsize=24)
plt.xlabel('Value', fontsize=14)
plt.ylabel('Square of Value', fontsize=14)

# 设置刻度标记的大小
# 这里的参数看不懂：both，which什么的...
plt.tick_params(axis='both', which='major', labelsize=14)

plt.show()

# 画出来的图和教材里的不一样，特别是原图x轴保留三位小数，这里两位，与y轴不同，并且在端点处没有标记数字，而原图有
```

### 15.2.4 使用scatter()绘制一系列点
    向scatter()传递两个分别包含x值和y值的列表
    matplotlib依次从每个列表中读取一个值来绘制一个点，如下共五个点。

```python
# scatter_squares.py
import matplotlib.pyplot as plt

# 15.2.4新增：
x_values = [1, 2, 3, 4, 5]
y_values = [1, 4, 9, 16, 25]

# plt.scatter(2, 4, s=200)  # s设置点的尺寸
# 15.2.4修改
plt.scatter(x_values, y_values, s=100)

# 设置图表标题并给坐标轴加上标签
plt.title('Square Number', fontsize=24)
plt.xlabel('Value', fontsize=14)
plt.ylabel('Square of Value', fontsize=14)

# 设置刻度标记的大小
# 这里的参数看不懂：both，which什么的...
plt.tick_params(axis='both', which='major', labelsize=14)

plt.show()

# 画出来的图和教材里的不一样，特别是原图x轴保留三位小数，这里两位，与y轴不同，并且在端点处没有标记数字，而原图有
```
