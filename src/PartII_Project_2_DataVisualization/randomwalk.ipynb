### 15.3.4 模拟多次随机漫步
    一种办法：将代码放入while循环中

```python
# rw_visual.py
import matplotlib.pyplot as plt
from random_walk import RandomWalk

# 15.3.4新增：加入while循环来模拟多次漫步
while True:
    # 创建一个RandomWalk实例，并将其包含的点都绘制出来
    rw = RandomWalk()
    rw.fill_walk()
    plt.scatter(rw.x_values, rw.y_values, s=15)
    plt.show()
    
    # 15.3.4新增：循环停止的判断
    keep_running = input('Make another walk?(y/n): ')
    if keep_running == 'n':
        break
```

### 15.3.6 给点着色
    使用颜色映射来指出漫步中各点的先后顺序，并删除每个点的黑色轮廓
    先后顺序着色：传递参数c，并将其设置为一个列表，其中包含各点的先后顺序

```python
# rw_visual.py
import matplotlib.pyplot as plt
from random_walk import RandomWalk

# 15.3.4新增：加入while循环来模拟多次漫步
while True:
    # 创建一个RandomWalk实例，并将其包含的点都绘制出来
    rw = RandomWalk()
    rw.fill_walk()
    
    # 15.3.6新增：新增点数变量，并在scatter中加入参数c，cmap，edgecolor
    point_numbers = list(range(rw.num_points))
    plt.scatter(rw.x_values, rw.y_values, c=point_numbers, cmap=plt.cm.Blues,
                edgecolor='none', s=15)
    plt.show()
    
    # 15.3.4新增：循环停止的判断
    keep_running = input('Make another walk?(y/n): ')
    if keep_running == 'n':
        break
```

### 15.3.7 重新绘制起点和终点
    让起点和终点变得更大，并显示不同的颜色以突出。

```python
# rw_visual.py
import matplotlib.pyplot as plt
from random_walk import RandomWalk

# 15.3.4新增：加入while循环来模拟多次漫步
while True:
    # 创建一个RandomWalk实例，并将其包含的点都绘制出来
    rw = RandomWalk()
    rw.fill_walk()
    
    # 15.3.6新增：新增点数变量，并在scatter中加入参数c，cmap，edgecolor
    point_numbers = list(range(rw.num_points))
        # 用range生成一个数字列表，其中包含的数字个数与漫步包含的点数相同
        # 将这个列表存储在变量中，以便后面使用它来设置每个漫步点的颜色
    plt.scatter(rw.x_values, rw.y_values, c=point_numbers, cmap=plt.cm.Blues,
                edgecolor='none', s=15)
    
    # 15.3.7新增：两个scatter，分别表示起点和终点，以突出(不同颜色和大小)
    plt.scatter(0,0,c='green',edgecolors='none',s=100) # 起点
    plt.scatter(rw.x_values[-1],rw.y_values[-1],c='red',edgecolor='none',
                s=100) # 终点
    
    plt.show()
    
    # 15.3.4新增：循环停止的判断
    keep_running = input('Make another walk?(y/n): ')
    if keep_running == 'n':
        break
```

### 15.3.8 隐藏坐标轴

```python
# rw_visual.py
import matplotlib.pyplot as plt
from random_walk import RandomWalk

# 15.3.4新增：加入while循环来模拟多次漫步
while True:
    # 创建一个RandomWalk实例，并将其包含的点都绘制出来
    rw = RandomWalk()
    rw.fill_walk()
    
    # 15.3.6新增：新增点数变量，并在scatter中加入参数c，cmap，edgecolor
    point_numbers = list(range(rw.num_points))
        # 用range生成一个数字列表，其中包含的数字个数与漫步包含的点数相同
        # 将这个列表存储在变量中，以便后面使用它来设置每个漫步点的颜色
    plt.scatter(rw.x_values, rw.y_values, c=point_numbers, cmap=plt.cm.Blues,
                edgecolor='none', s=15)
    
    # 15.3.7新增：两个scatter，分别表示起点和终点，以突出(不同颜色和大小)
    plt.scatter(0,0,c='green',edgecolors='none',s=100) # 起点
    plt.scatter(rw.x_values[-1],rw.y_values[-1],c='red',edgecolor='none',
                s=100) # 终点
    
    # 15.3.8新增：隐藏坐标轴
    plt.axes().get_xaxis().set_visible(False)
    plt.axes().get_yaxis().set_visible(False)
    
    plt.show()
    
    # 15.3.4新增：循环停止的判断
    keep_running = input('Make another walk?(y/n): ')
    if keep_running == 'n':
        break
```

### 15.3.9 增加点数
    创建RandomWalk实例时，增大num_points，并调整每个点的大小
